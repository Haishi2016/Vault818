# Percepron and LTU

## Definition

The Perceptron is one of hte simplest ANN architectures. In the simplest form, it uses a single Linear Threshold (LTU) that is defined as:

$h_{\mathbf{w}}(\mathbf{x}) = step(z) = step(\mathbf{w}^t\cdot\mathbf{x}) = step(w_{1}x_{1} + w_{2}x_{2} + \dots + w_{n}x_{n})$

Where the step function can be a _heaviside_ function

$heaviside(z) = \begin{cases} 0 &\text{if } z < 0 \\ 1 &\text{if } z \geq 0 \end{cases}$

Or a _sign_ function

$sign(z) = \begin{cases} -1 &\text{if } z < 0 \\ 0 &\text{if } z = 0 \\ 1 &\text{if } z > 0 \end{cases}$

A more complex Perceptron (network) is formed by a single _output_ layer of LTUs that are fully connected to an _input layer_ that is made up by input neurons taht simply ooutputs what it receives. A bias nuron, which constantly outputs 1, can be added to the input layer as a _bias neuron_.

## Percepron learning rule

A perceptron network learns by reducing the weight of neurons who have led to the wrong output:

$w_{i,j}^{(next step)} = w_{i,j} + \eta(\hat{y_{j}}-y_{j})x_{i}$

## Use Perceptron to classify Iris dataset

The following network uses a single LTU network to separte Iris Setosa from other Iris types.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

iris = load_iris()
X = iris.data[:, (2,3)] # petal Length, petal width
y = (iris.target == 0).astype(np.int)

per_clf = Perceptron(max_iter=100, tol=-np.infty, random_state=42)
per_clf.fit(X, y)
```

```python
y_pred = per_clf.predict([[2, 0.5]])
print(y_pred)
```

```python
print(per_clf.coef_)
print(per_clf.intercept_)
```

```python
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap

a = -per_clf.coef_[0][0] / per_clf.coef_[0][1]
b = -per_clf.intercept_ / per_clf.coef_[0][1]

axes = [0, 5, 0, 2]

x0, x1 = np.meshgrid(
        np.linspace(axes[0], axes[1], 500).reshape(-1, 1),
        np.linspace(axes[2], axes[3], 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]
y_predict = per_clf.predict(X_new)
zz = y_predict.reshape(x0.shape)

plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs", label="Not Iris-Setosa")
plt.plot(X[y==1, 0], X[y==1, 1], "yo", label="Iris-Setosa")

plt.plot([axes[0], axes[1]], [a * axes[0] + b, a * axes[1] + b], "k-", linewidth=3)

custom_cmap = ListedColormap(['#9898ff', '#fafab0'])

plt.contourf(x0, x1, zz, cmap=custom_cmap)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="lower right", fontsize=14)
plt.axis(axes)
plt.show()
```

## Multi-Layer Perceptron (MLP)

MLP can be used to overcome some of the limitations of Perceptrons. The layers between the input layer and the output layer are called _hidden layers_. Every laer except the output layer has an extra bias neuron that is fully connected to the next layer. When an ANN has multiple layers its called DNN (Deep Neural Network). MLP is trained by backpropagation, in which a forward pass is used to make prediction, and the measured errors are propergated backwards along the network to adjust weights.

The following network classifies MNIST dataset using MLP.

```python
import tensorflow as tf
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.astype(np.float32).reshape(-1, 28 * 28)/255.0
X_test = X_test.astype(np.float32).reshape(-1, 28*28)/255.0
y_train = y_train.astype(np.int32)
y_test = y_test.astype(np.int32)
X_valid, X_train = X_train[:5000], X_train[5000:]
y_valid, y_train = y_train[:5000], y_train[5000:]
```

```python
feature_cols = [tf.feature_column.numeric_column("X", shape=[28*28])]
dnn_clf = tf.estimator.DNNClassifier(hidden_units=[300,100], n_classes=10, feature_columns=feature_cols)
input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"X": X_train}, y=y_train, num_epochs=40, batch_size=50, shuffle=True)
dnn_clf.train(input_fn=input_fn)
```

```python
test_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"X": X_test}, y=y_test, shuffle=False)
eval_results = dnn_clf.evaluate(input_fn=test_input_fn)
```
